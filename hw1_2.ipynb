### 5. Undergraduate GPA(CGPA) (out of 10)
### 6. Research Experience(either 0 or 1)
### 7. Chance of Admit(ranging from 0 to 1)

### Plot HeatMap using seaborn
* Correlation between numeric features(GRE Score, TOEFL Score, University Rating, SOP, LOR, CGPA, Chance of Admit)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd_data = pd.read_csv('Admission_Predict_Ver1.1.csv')
pd_data = pd_data.drop(['Research'], axis=1)
```

```python
plt.figure(figsize=(12,10))
foo = sns.heatmap(pd_data.corr(), vmax=0.6, square=True, annot=True)
```

### Top-10 CGPA
* Sort CGPA from high to low
* Show their corresponding Chance of Admit

```python
pd_data = pd.read_csv('Admission_Predict_Ver1.1.csv')
```

```python
sort_by_cgpa = pd_data.sort_values('CGPA',ascending=False)
sort_by_cgpa = sort_by_cgpa.drop(['Serial No.', 'GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'Research', 'LOR '], axis = 1).head(10).reset_index()
sort_by_cgpa = sort_by_cgpa.drop(['index'], axis=1)
sort_by_cgpa
```

### Plot University Rating according to the top 50 GRE Score

```python
pd_data = pd.read_csv('Admission_Predict_Ver1.1.csv')
sort_by_cgpa = pd_data.sort_values('GRE Score',ascending=False)
sort_by_cgpa = sort_by_cgpa.drop(['Serial No.', 'CGPA', 'TOEFL Score', 'Chance of Admit ', 'SOP', 'Research', 'LOR '], axis = 1).head(50).reset_index()
plt.hist(sort_by_cgpa['University Rating'])
plt.grid(True)
```

### Stripplot University Rating according to the  CGPA

```python
pd_data = pd.read_csv('Admission_Predict_Ver1.1.csv')
sns.stripplot(pd_data['University Rating'], pd_data['CGPA'], jitter=True)
```

### TOEFL scores more than 100 and it corresponding Chance of Admit

```python
pd_data = pd.read_csv('Admission_Predict_Ver1.1.csv')
pd_data = pd_data[pd_data['TOEFL Score'] > 100]
pd_data.plot.scatter(x='TOEFL Score', y='Chance of Admit ')
```
